```python
# Making a face pixel art
import numpy as np
import matplotlib.pyplot as plt

art = np.zeros((10,10,3), dtype=int) + 255 
art[:, :] = [255, 255, 0]  
art[2:4, 2:4] = [0, 0, 0]
art[2:4, 6:8] = [0, 0, 0]
art[6:8, 3:7] = [255, 0, 0] 
plt.imshow(art)
plt.axis("off")
plt.show()
```

```python
# Making a heart 
import numpy as np
import matplotlib.pyplot as plt

art = np.zeros((10, 10, 3), dtype=int) + 255
heartCords = [(1,1), (1,2), (1, 3), (1,5), (1,6), (1,7), (2, 0), (2, 1), (2,2), (2,3), (2,4), (2,5), (2,6), (2,7), (2,8)]
for r, c in heartCords:
    art[r, c] = [255, 0, 0]
plt.imshow(art)
plt.axis("off")
plt.show()
```
