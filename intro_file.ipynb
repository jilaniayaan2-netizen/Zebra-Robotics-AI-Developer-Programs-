```python

import numpy as np

# Create a NumPy array
arr = np.array([1, 2, 3, 4, 5])
print("Original array:", arr)

# Basic operations
print("Add 5 to each element:", arr + 5)
print("Subtract 2 from each element:", arr - 2)
print("Multiply each element by 3:", arr * 3)
print("Divide each element by 2:", arr / 2)
print("Square each element:", arr ** 2)

# Element-wise operations
arr2 = np.array([6, 7, 8, 9, 10])
print("Array 2:", arr2)
print("Addition of arrays:", arr + arr2)
print("Multiplication of arrays:", arr * arr2)

# Aggregate functions
print("Sum of array:", np.sum(arr))
print("Mean of array:", np.mean(arr))
print("Maximum element:", np.max(arr))
print("Minimum element:", np.min(arr))

# Array indexing and slicing
print("First element:", arr[0])
print("Last element:", arr[-1])
print("Elements from index 2 to 4:", arr[2:5])

# Reshaping arrays
reshaped_arr = arr.reshape(5, 1)
print("Reshaped array:\n", reshaped_arr)

# Creating arrays with specific values
zeros_arr = np.zeros((2, 3))
print("Array of zeros:\n", zeros_arr)

ones_arr = np.ones((3, 2))
print("Array of ones:\n", ones_arr)

range_arr = np.arange(0, 10, 2)
print("Array with a range of values:", range_arr)
```

# 4. Operators

Operators in Programming are special symbols that help computers perform different tasks:

1. Arithmetic Operators (+, -, *, /, %) are used for basic math calculations.
2. Comparison Operators (==, !=, >, <, >=, <=) are used to compare values.
3. Compound Operators (+=, -=, *=, /=) provide shortcuts for performing operations and assigning values.
Operators make it easier for programmers to manipulate data and control a program's behavior.


Your task is to use these operators and perform operation on data or variables in the code cell given below

```python
x = 25
y = 35 
print(x+y)
print(y-x)
print(x*y)
print(x/y)
print(x%y)

if(x <= y):
    y+=5
    print(y)

if(y >= x):
    y-=15
    print(y)

if(y == x):
    x*=10
    print(x)

if(x != y):
    x/=5
    print(x)
```

# 5. String Operations

Define a string and then perform the following opertions
1. capitalize() - Converts the first character to upper case
2. count() - Returns the number of times a specified value occurs in a string
3. format() - Formats specified values in a string
4. find() - Searches the string for a specified value and returns the position of where it was found
5. lower() - Converts a string into lower case
6. split() - Splits the string at the specified separator, and returns a list
7. strip() - Returns a trimmed version of the string
8. upper() - Converts a string into upper case

```python
word = "helloWorld"

print(word.capitalize())
print(word.count("l"))
print(word.format("h"))
print(word.find("o"))
print(word.lower())
print(word.split("W"))
print(word.strip())
print(word.upper())
```

# 7. Loops

There are 2 types of loops in python - while loop and for loop

while loop - it works with a condition and keeps running untill the condition is true

for loop - it works with itreation and run the code untill it reaches the last element of the variable it is iterating

in the code cell given below make a code to print your name 10 times using for loop and while loop

```python
numOfLoops = 10
x = 10 

while(numOfLoops >= 0):
    print("Ayaan")
    numOfLoops -= 1

for x in range(x):
    print("Ayaan")
    x -= 1
```

Make a code to check if the input number is prime or not

```python
userNumber = int(input("Enter a number: "))

for x in range(2, userNumber):
    if(userNumber%x == 0): 
        print(userNumber, "is not a prime number.")
        break
    else: 
        print(userNumber, "is a prime number.")
        break
```
