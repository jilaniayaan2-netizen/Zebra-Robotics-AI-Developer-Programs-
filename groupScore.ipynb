```python
import statistics as stats 

groupXScores = [78, 80, 79, 77, 81]
groupYScores = [60, 85, 55, 90, 75]
groupXVariance = stats.variance(groupXScores)
groupYVariance = stats.variance(groupYScores)
groupXStDev = stats.stdev(groupXScores)
groupYStDev = stats.stdev(groupYScores)
groupXRange = max(groupXScores) - min(groupXScores)
groupYRange = max(groupYScores) - min(groupYScores)

print("The variance of student group X is", groupXVariance)
print("The variance of student group Y is", groupYVariance)
print("The standard deviation of student group X is", groupXStDev)
print("The standard deviation of student group Y is", groupYStDev)
print("The range of student group X is", groupXRange)
print("The range of student group Y is", groupYRange)
```
