Write a function to find the even numbers in a list and return them as an output

```python
listOfNum = [1, 2, 3, 4, 5, 6]
def evenOrNot(listOfNum):
    x = len(listOfNum)
    while x > 0:
        if(listOfNum[x-1]%2 == 0):
            print(x, "is even.")
        else:
            print(x, "is odd.")
        x -= 1

evenOrNot(listOfNum)
```

# 9. Classes and objects

Object-Oriented Programming (OOP) is the process of creating your own variable type. This variable type allows us to represent abstract concepts and give them attributes and methods to use them.

Objects are an encapsulation of variables and functions into a single entity. Objects get their variables and functions from classes. Classes are essentially a template to create your objects.

There is an example below

```python
class Test_class:  ## defining the class
    var = 'hello'

    def name_func(self):     # defining a method inside a class
        print('this is a test function in a class')

my_obj = Test_class()   # assigning a class to the object

print(my_obj.var)
my_obj.name_func()
```

# Dictionaries

A dictionary is a data type similar to arrays, but works with keys and values instead of indexes. Each value stored in a dictionary can be accessed using a key, which is any type of object (a string, a number, a list, etc.) instead of using its index to address it.

Create a dictionary with phone numbers. Then perform various dictionary operations on it and explain them in the class

```python

phone_numbers = {
    "Alice": "555-123-4567",
    "Bob": "555-987-6543",
    "Charlie": "555-111-2222"
}

# Accessing a value using a key
print(phone_numbers["Alice"])  # Output: 555-123-4567

# Adding a new key-value pair
phone_numbers["David"] = "555-333-4444"
print(phone_numbers)

# Modifying an existing value
phone_numbers["Bob"] = "555-555-5555"
print(phone_numbers)

# Checking if a key exists
if "Eve" in phone_numbers:
    print(phone_numbers["Eve"])
else:
    print("Eve's phone number not found")

# Removing a key-value pair
del phone_numbers["Charlie"]
print(phone_numbers)

# Getting all keys
print(phone_numbers.keys())

# Getting all values
print(phone_numbers.values())

# Iterating through the dictionary
for name, number in phone_numbers.items():
    print(f"{name}'s phone number is {number}")
```

# File Access

File handling is a fundamental aspect of Python programming, enabling you to create, read, update, and delete filesâ€”commonly referred to as CRUD operations. Here's a concise guide to performing these operations:

The key function for working with files in Python is the open() function.

The open() function takes two parameters; filename, and mode.

There are four different methods (modes) for opening a file:

"r" - Read - Default value. Opens a file for reading, error if the file does not exist

"a" - Append - Opens a file for appending, creates the file if it does not exist

"w" - Write - Opens a file for writing, creates the file if it does not exist

"x" - Create - Creates the specified file, returns an error if the file exists

### Creating or opening a file in write mode

```python
# Define the filename
filename = "example.txt"

# 1. Write to a file (this will create the file or overwrite if it exists)
file = open(filename, "w")
file.write("Hello, world!\n")
file.write("This is the first line.\n")
```

### Creating or Opening a file in the append mode -  in this mode we can open a text file and then write in it

```python
# 2. Append to the file (adds content to the end without overwriting)
filename = "append.txt"
file =  open(filename, "a")
file.write("This line is appended.\n")
file.write("HELLO WORLD.")
```

### Opening the file in read format to take out the content and use it

```python
# 3. Read the entire content of the file
file = open(filename, "r")
content = file.read()
print("Full content of the file:")
print(content)
```

### Code to delete a file

```python
import os

# Specify the path to the file
file_path = "example.txt"

# Check if the file exists before attempting to delete it
if os.path.exists(file_path):
    os.remove(file_path)
    print(f"File '{file_path}' has been deleted successfully.")
else:
    print(f"The file '{file_path}' does not exist.")
```
