```python
import numpy as np  

# Step 1: Generate the Data
randArr = np.array(np.random.randint(0, 100))
for x in range(29):
    randArr = np.append(randArr, np.random.randint(0, 100))

# Step 2: Reshape It
newArr = randArr.reshape(5, 6)

# Step 3: Iterate and Display
for x in range(0,5):
    print("Array 1, row", x+1 ,newArr[x,0:5]) 

# Step 4: Join Arrays
randArr2 = np.array(np.random.randint(0, 100))
for x in range(29):
    randArr2 = np.append(randArr2, np.random.randint(0, 100))
newArr2 = randArr2.reshape(5, 6)
joinedArr = np.concatenate((newArr, newArr2), axis=1)

# Step 5: Split Arrays
split = np.hsplit(joinedArr, 2)
splitArr = np.array(split)
splitArr1 = np.array(splitArr[0])
splitArr2 = np.array(splitArr[1])
splitArr1 = splitArr1.reshape(5, 6)
splitArr2 = splitArr2.reshape(5, 6)

# Step 6: Search
y = np.where(randArr > 50)
print("The first array is greater than 50 at the indices", y)
print("The values in the array that are greater than 50 are", randArr[y])

# Step 7: Sort
print("Row 1 sorted looks like", np.sort(newArr[0]))
print("Row 2 sorted looks like", np.sort(newArr[1]))
print("Row 3 sorted looks like", np.sort(newArr[2]))
print("Row 4 sorted looks like", np.sort(newArr[3]))
print("Row 5 sorted looks like", np.sort(newArr[4]))

# Step 8: Filter
filterArr = []
for element in randArr:
    if element%2 == 0:
        filterArr.append(True)
    else:
        filterArr.append(False)
randArr = randArr[filterArr]
print("A filtered array that only has even numbers looks like", randArr)
```
