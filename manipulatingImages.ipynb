```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load Image
img = Image.open("car.jpg")
img_arr = np.array(img)
print("Image shape:", img_arr.shape)
plt.imshow(img_arr)
plt.axis("off")
plt.show()

# Make the image red-tinted
red_img = img_arr.copy()
red_img[:, :, 1] = 0   # remove green
red_img[:, :, 2] = 0   # remove blue
plt.imshow(red_img)
plt.axis("off")
plt.show()

# Make the image green-tinted
green_img = img_arr.copy()
green_img[:, :, 0] = 0   # remove red
green_img[:, :, 2] = 0   # remove blue
plt.imshow(green_img)
plt.axis("off")
plt.show()

# Make the image blue-tinted
green_img = img_arr.copy()
green_img[:, :, 0] = 0   # remove red
green_img[:, :, 1] = 0   # remove green
plt.imshow(green_img)
plt.axis("off")
plt.show()

# Double size (stretch with np.repeat)
big_img = np.repeat(np.repeat(img_arr, 2, axis= 0), 2, axis=1)
plt.imshow(big_img)
plt.axis("off")
plt.show()

# Crop the middle part of the image
h, w, _ = img_arr.shape
crop_img = img_arr[h//3:h//3*2, w//3:w//3*2]  # middle square
crop_img2 = img_arr[h//4:h//4*3, w//4:w//4*3] # slightly zoomed out middle sqaure
crop_img3 = img_arr[h//2:h//2*2, w//2:w//2*2] # bottom right
plt.imshow(crop_img)
plt.axis("off")
plt.show()
plt.imshow(crop_img2)
plt.axis("off")
plt.show()
plt.imshow(crop_img3)
plt.axis("off")
plt.show()

# Flip horizontally
flip_img = np.fliplr(img_arr)
plt.imshow(flip_img)
plt.axis("off")
plt.show()
 
# Rotate 90 degrees
rot_img = np.rot90(img_arr)
plt.imshow(rot_img)
plt.axis("off")
plt.show()
```

```python
# Convert to black and white
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

img = Image.open("car2.jpg")
img_arr = np.array(img)
plt.imshow(img)
plt.axis("off")
plt.show()

def showGreyscale(img):
    grey = img_arr.mean(axis=2)
    gi = np.stack([grey, grey, grey], axis=2)
    plt.imshow(gi.astype(int))
    plt.axis("off")
    plt.show()

showGreyscale(img)
```
